##### Problem with the above structure
  - Different analysis tools demand different setups. 
  For example: 
  - Statistical calculations and correlation table can have plenty of variables
  - The correlation plot can only have two variables. 
  - The multi-plot needs to devide the variables among $y$-axes, and it would be hard to read it if there are more than two $y$-axes and say 5-6 variables.

#### &check;1.1 
  
##### A more logical way would be:
  - each tool of a batch points out the variables that are part of that plot
  - to let 'multi_plot' points out a list of y-axises of the plot depending on the unit, and the variables therin 
  - use a flag 'all', when all variables of the group is to be part of the tool, and then use the order specified within the group, e.g. as given by the list of the dict-key '_group_vars'
  
##### Moreover: 
 - There is a limit on the correlation plot is that it can only take *two variables*. 
    - However, in the batch job '7var' of the group 'abisko 8 var', both 3 and 4 use 'corr-plot'.
    - The number of corr-plots needed for 5 variables is 10, for 6 variables it is 15 (for $k$ variables it is "*k-choose-2*" $k(k-1)/2$). 
 - The 'all' flag can only be allowed for the tools 'statistics' and possibly the 'corr-table'.
 - For multiplot there should be a limit on the number of:
      - variables (say 6) 
      - the number of y-axises (say 2 or 3) <br>
   Possibly we could automatically create a new multi-plot when a limit is reached
