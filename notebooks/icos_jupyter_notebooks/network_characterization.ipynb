# Characterization of networks of atmospheric measurement stations

### Background

ICOS currently has 13 member states and deploys almost 150 stations in the three domains atmosphere, ecosystem and ocean. Essential to monitor the net greenhouse gas balance of Europe and its immediate surroundings is to optimise the observation network to the evolving needs of science and policy makers. Providing the necessary information to further improve and develop models that answer the needs of these users is crucial. Extending the network to new member states and optimising the density in existing member states to fulfil these requirements can be assisted by network design tools.


For the atmosphere domain the current ICOS network consists of 38 stations that monitor CO2, CH4, CO and in increasing cases also N2O with high precision and accuracy, presenting an unprecedented dense network compared to the rest of the world. One of the main applications of the ICOS atmosphere observations is the use in regional inversion models. Inversion models, by combining observations, prior information and models, can be used to reduce the uncertainty on estimates of the net greenhouse gas fluxes at high spatial and temporal resolution and to distinguish contributions of different source and sink categories. Previous studies show that uncertainty in the atmospheric transport as well as the still existing regional gaps in the observation network are the determining factors that currently limits these inverse models. Future enhancement of methods and models and increasing high quality observations can reduce these limitations. The location for additional (ICOS-) stations needs to be carefully planned to create maximum impact on the models and enhancement of the existing network.

### Tool overview

To assist the national networks in evaluating the current and possible atmospheric network configurations, the ICOS Carbon Portal has extended its STILT footprint service with a Jupyter notebook based tool. The tool allows the user to test any potential network configuration on its capability to capture the influence of sources and sinks. It does so by exploring the characteristics of combined influence (footprint) areas. The resulting combined footprint of the user’s selection is displayed in three different types of maps including one that simply shows the combined footprint and two maps that are based on the absolute coverage of the footprints (referred to as “footprint masks”). In a next step, the footprint mask coverage is examined in terms of the extent to which it covers individual countries. The results are summarized in a table. 

In a final step, the footprints are multiplied by underlying land cover and population data to get a better understanding of what the representation of the individual countries encompasses. The user has two options for how this should be done: One option is to use the footprint mask which means that the land cover areas and population underlying any part of the footprint mask are summarized (area of the different land cover types and total population count). The alternative is to multiply the land cover area and population count with a sensitivity weighted footprint. In case of a network configuration including more than one site, the highest sensitivity value for each area (cell) in the domain is used. For comparison between the represented countries the results are displayed in one common table showing all user-selected countries.

#### Defining an influence area using footprints

The extent and shape of the resulting footprints depend on what percentage of the individual stations’ footprints should be used. This choice is up to the user and should be carefully considered. For this consideration, an understanding of how the footprints are generated is crucial:

Footprints used in this tool are generated using a the lagrangian atmospheric transport model STILT which is implemented as an <a href="https://www.icos-cp.eu/data-services/tools/stilt-footprint" target="blank">online tool</a> at the ICOS Carbon Portal. Output footprints are presented on a grid with 1/12×1/8 degrees cells (approximately 10km x 10km) where the cell values represent the cell area’s estimated surface influence (“sensitivity”) in ppm / (μmol/ (m²s)) on the atmospheric tracer concentration at the station.

Individual footprints are generated every three hours (between 0:00 and 21:00 UTC) and are based on a 10-days backward simulation. To represent where the influences on the concentrations measured at the stations generally come from, average footprints for an entire year are used. The year 2018 is used in this tool. The 10 days’ backward simulation means that a footprint can have a very large extent. When averaging many footprints almost the entire STILT model domain (Europe) will have influenced the concentration at the station at some point in time. To limit the coverage to areas with more significant influence, the user specifies <b>what percentage of the average footprints should be used</b>. This is achieved by including the sensitivity values of the footprint cells in descending order until the specified percentage of the total sensitivity is reached.

#### Land cover breakdown and population

<a href=https://land.copernicus.eu/pan-european/corine-land-cover/clc2018 target="blank">CORINE (2018) land cover data</a> is multiplied by the footprints to estimate the land cover breakdowns of the different countries they cover. Multiplication by the combined footprint results in a weighing of the area of each land cover class within the footprint by the sensitivity to each cell (km² area* (ppm / (μmol/ (m²s)). Multiplication by the combined footprint mask results in a simple estimate of the area of each land cover class within the footprint. The same principles of multiplication with the two footprint types are used to estimate the influence and size of the population that lives within the influence area. The population data used come from <a href=”https://ec.europa.eu/eurostat/web/gisco/geodata/reference-data/population-distribution-demography/geostat” target=”blank”>GEOSTAT (2018)</a>.


The user specifies which countries should be considered. If all countries are selected, the results for those covered by the footprints will be shown in the output graphs. Furthermore, the user selects whether the exclusive economic zone (EEZ) should be considered part of the country. If a country does not have an EEZ, the option to include it will not affect the output for that specific country. 

#### Adding stations to the network
In order to optimize the placements of future stations and station networks, the tool allows for a comparison between the specified network (referred to as “base network”) and the same network but with one or more additional stations included (referred to as “comparison network”). Stations do not necessarily need to exist: STILT footprints can be generated for any point(s) and will appear in the dropdown as soon as all footprints for 2018 <a href= ”https://www.icos-cp.eu/about-stilt-calculator” target=”blank”>have been computed</a>.
