# Ecosystem Site Anomaly Visualization

### Introduction
This notebook allows users to visualize anomalies in ecosystem variables. Daily averages for a selected year are compared to daily averages for a reference period of either 11 or 21 years. The results are aggregated to either months of the year, or kept as days of the year in which case all data for a selected month is plotted. The data can appear anomalous for different reasons, including anomalous weather, crop rotation, or other land use changes.<br>

#### Notebook Citation

Storm, I., Klasen, V., 2023. Ecosystem Site Anomaly Visualization Notebook Tool. ICOS ERIC - Carbon Portal. https://doi.org/10.18160/0GP0-HW10

### Flux data 

Eddy covariance fluxes at 73 sites that are part of the collection "<a href="https://www.icos-cp.eu/data-products/2G60-ZHAK" target="_blank">Warm Winter 2020</a>" have been considered for these analyses. 

The daily average values found in the files with the structure "<b>FLX_</b>SE-Deg<b>\_FLUXNET2015_FULLSET_DD_</b>2001<b>-2020_beta-3</b>" are used.

In 2023, the tool was extended with data from the "<a href = "https://www.icos-cp.eu/data-products/ecosystem-release" target="_blank">ICOS release 2023-1 of Level 2 Ecosystem data</a>" for the years 2021 and 2022. These years are available for a subset of sites in the tool, including DE-Kli, FR-Aur, CZ-wet, BE-Lon, DE-Gri, DE-Hai, SE-Deg, IT-Tor, FI-Let, FR-Fon, and FI-Hyy.

The daily average values found in the files with the structure "<b>ICOSETC_</b>SE-Deg<b>\_FLUXNET_DD_L2</b>" are used.

Considered variables, and their associated quality control (QC) flags, include:

- GPP_DT_VUT_REF: Gross Primary Production during daytime in µmol/m²/s. 
    - NEE_VUT_REF_QC (as the quality of all the fluxes derived from NEE, Net Ecosystem Exchange, primarily depends on the quality of the original fluxes).
- SW_IN_F: Shortwave incoming radiation in W/m². 
    - SW_IN_F_QC
- VPD_F: Vapor pressure deficit in hPa.
    - VPD_F_QC

The quality control flags indicate the percentage of high-quality half-hourly data used to calculate the daily value. To exclude data of poor quality, a threshold of 70% (0.7) was implemented.

### Flux data analysis

Two reference periods with daily averages averaged for years 2000-2020 (21 years) and 2010-2020 (11 years) are available for all feasible sites, i.e. sites with available data for the whole period. The averages over a selected reference period for selected sites are compared to daily averages for selected years. 

Intended use cases include:

- Visualize anomalies in the same variable at two sites for the same year.
- Visualize anomalies in the same variable at the same site for two different years.
- Visualize anomalies in two different variables at the same site for the same year. 

### Output

Below is an example for anomalies in the same variable at two sites for the same year. Note that the two types of graphs, the circular bar charts and the more conventional horizontal bar chart show the same information. 

The example is showing daytime GPP at Hainich (DE-Hai) in Germany and Hyytiala (FI-Hyy) in Finland during the drought year of 2018. We can see that the two sites show quite different responses. Both start out with positive anomalies in the spring but for Hainich, large negative anomalies are present during the summer and autumn months. The anomalies at Hyytiala during those months are not significant with respect to the years 2000-2020 (data for the year 2018 is not excluded from the reference period average).

Note that there is no valid data (QC < 0.7) available from January 1 through April 15 at Hyytiala. While there is a bar showing the anomaly for the latter half of April, the specific days that are flagged become apparent when April is selected as a temporal filter (not shown).

### Visual Elements

#### Bars
The bars consist of the following elements:
* A grey bar showing the reference value, i.e. the mean of all relevant values from the chosen reference period, in the example 2000-2020.
* A slim darker grey bar showing the standard deviation within the reference values.
* A green bar showing the positive or negative anomaly of the value (2018 values in the example) with respect to the reference value.

Flagged daily values (QC < 0.7) are not included when the reference values and the standard deviation within the reference values are calculated. Within the context of a monthly temporal scope and the specified reference period (2010-2020), only 11 reference values are taken into account. If more than five of these values are excluded (more than 15 in case of the longer reference period), the standard deviation bar will not be displayed. An illustrative instance of this can be observed on November 28, 2010, at Hetzdorf (DE_Hzd). The output file, which accompanies the visual representations, contains columns indicating the number of values utilized in the calculation of daily and monthly reference values.  

<img src="ecosystem_site_anomaly_visualization/img/visual_explanation_bars.png" align="left" width="60%"> <br clear="all" />

#### Pie charts

In the middle of the circular bar charts, pie charts show the total plotted values. The bottom half circle represents the total reference values, with a dark green portion representing negative anomalies. The upper half circle represents the total 2018 values, with a light green portion representing positive anomalies.

The radii of the two circles are proportional to the sum of the plotted values and the angles of the green portions are proportional to the sum of the plotted anomalies.

In the given example, we can see that the total values in 2018 are lower than the reference values, since the radius of the upper circle is smaller than that of the bottom circle. We can also see that there were months with positive anomalies, as there is a light green portion in the upper circle, but the sum of all positive anomalies is small compared to that of the negative anomalies, as there is a dark green portion in the bottom circle with a bigger angle.
